```python
from langchain_core.tools import tool

# custom tool for multiplication
@tool
def multiply(a: int, b: int)->int:
  """Multiplying two number"""
  return a*b

@tool
def add(a: int, b: int)-> int:
  """Adding two numbers"""
  return a+b
```

```python
# Building it in a class
class MathToolkit:
  def get_tool(self):
    return [add, multiply]
```

```python
toolkit = MathToolkit() # object of class
tools = toolkit.get_tool() # Calling that classmethod

for tool in tools:
  print(tool.name, "=>", tool.description)
```

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
# makeing the schema
class BundleInput(BaseModel):
  a : int = Field(required = True, description = "First Number")
  b : int = Field(required = True, description = "Second Number")
```

```python
#Tools function
# custion
def multiply(a,b):
  return a*b

def add(a,b):
  return a+b

def sub(a,b):
  return a-b
```

```python
# creating the main tool object
add_tool = StructuredTool.from_function(
    func= add,
    name= "add",
    description= "Adding two number",
    args_schema = BundleInput
)

sub_tool = StructuredTool.from_function(
    func = sub,
    name ="subtraction",
    description = "Subtracting two numbers",
    args_schema = BundleInput
)

multiply_tool = StructuredTool.from_function(
    func = multiply,
    name = "Multiply",
    description = "Multiplying two numbers",
    args_schema = BundleInput
)
```

```python
# creating a class to bundle these
class Mathtoolkit:
  def get_tools(self):
    return [add_tool, multiply_tool, sub_tool]
```

```python
# calling the toolkit  using class object
toolkit = Mathtoolkit()
tool = toolkit.get_tools()

for tol in tool:
  print(tol.name, "=>", tol.description)
```
