```python
from langchain_openrouter import ChatOpenRouter
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_run = DuckDuckGoSearchRun()

# result = search_run.invoke("heartstopper forever")
```

```python
# result
```

```python
# we will create a llm
LLM = ChatOpenRouter(model = "cohere/north-mini-code:free")
# LLM.invoke("hi")
```

# now we have an LLM and a Tool now we can create a Agent out of it

# Out Agent will be a ReAct agent that is Reasoning + Action

```python
from langgraph.prebuilt import create_react_agent
from langchain_classic.agents import AgentExecutor
import langchainhub as hub
```

```python
# step 2: create a ReAct agent manually(ReAct is a design pattern)
Agent = create_react_agent(
    model = LLM,
    tools = [search_run]
)
#  this helps in planning and decision making
```

```python
# step 3: Wrap it with AgentExecutor
# agent_executor = AgentExecutor(
#     agent = Agent,
#     tools = [search_run],
#     verbose = True
# )
# executor helps in executing those plans tools and decisions made by the agent
```

```python
# step 4: lets invoke it
response = Agent.invoke(
    {
        "messages": [("user", "can you plan my 7 days trip to japan, everything you decide and just tell me the plan")]
    }
)
```
